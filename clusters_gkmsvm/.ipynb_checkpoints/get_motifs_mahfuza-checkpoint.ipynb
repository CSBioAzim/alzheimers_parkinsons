```python
def rolling_window(a, window_width):
    """Create a new array suitable for rolling window operation
    Adopted from http://www.rigtorp.se/2011/01/01/rolling-statistics-numpy.html,
    also discussed in this PR: https://github.com/numpy/numpy/issues/7753
    Args:
      a: input array of shape (..., positions)
      window_width: width of the window to scan
    Returns:
      array of shape (..., positions - window_width + 1, window_width)
    """
    shape = a.shape[:-1] + (a.shape[-1] - window_width + 1, window_width)
    strides = a.strides + (a.strides[-1],)
    return np.lib.stride_tricks.as_strided(a, shape=shape, strides=strides)
def sliding_sum(query, target_scores):
    query = query.swapaxes(-2, -1)  #
    target_scores = target_scores.swapaxes(-2, -1)
    assert target_scores.shape[-1] >= query.shape[-1]  # target needs to be longer than the query
    window_len = query.shape[-1]
    ta_strided = rolling_window(target_scores, window_len).swapaxes(-2, -1)
    ta_L1_norm = np.sum((ta_strided), axis=(-3, -2))
    return ta_L1_norm
def load_deeplift_data(deeplift_hdf, keys=['scores', 'one_hot']):
    #deeplift_hdf = "{0}/{1}/Naive_deeplift2019/task_273-naivegw/summit.h5".format(srv_root, data_name)
    fp = h5py.File(deeplift_hdf, "r")
    hyp_scores = fp['deeplift_scores'][:]
    one_hot = fp['inputs'][:]
    shuffled_onehot = fp['shuffled_inputs'][:]
    df = OrderedDict()
    for key in list(fp['metadata/range']):
        df[key] = fp['metadata/range/{}'.format(key)][:]
    df = pd.DataFrame(df)
    df.chr = np.array([df.chr.values[i].decode('utf-8') for i in range(df.shape[0])])
    deeplift_data = OrderedDict()
    if 'one_hot' in keys:
        deeplift_data['one_hot'] = one_hot
    if 'shuffled_onehot' in keys:
        deeplift_data['shuffled_onehot'] = shuffled_onehot[:, 0, :, :]
    if 'peaks' in keys :
        deeplift_data['peaks'] = df
    if 'hyp_scores' in keys :
        deeplift_data['hyp_scores'] = hyp_scores
    if 'scores' in keys :
        deeplift_data['scores'] = np.multiply(hyp_scores, one_hot)#no need to sum before mult, taken care of in deeplift, not in deepshap though
    if 'shuffled_scores' in keys :
        #deeplift_data['shuffled_scores'] = np.sum(np.multiply(hyp_scores, shuffled_onehot[:, 0, :, :]), axis=2) ##1st set of shuffles among 10sets
        deeplift_data['shuffled_scores'] = np.multiply(hyp_scores, shuffled_onehot[:, 0, :, :])
    return deeplift_data
deeplift_hdf = "/mnt/lab_data2/msharmin/oc-atlas/deeplift2019_h5/task_224-naivegw/summit.h5"
deeplift_data = load_deeplift_data(deeplift_hdf, keys=['one_hot', 'scores', 'shuffled_scores',
                                                                   'peaks'])
scores = deeplift_data['scores']
hyp_scores = deeplift_data['shuffled_scores']
#scores[scores<=0]=0
#hyp_scores[hyp_scores<=0]=0
#average_motif_len = 15
#a = sliding_sum(np.ones((average_motif_len, 4)), hyp_scores)
#a = a.flatten()/average_motif_len
```

```python
scores_ = [np.sum(j) for j in scores[0][300:700]]
shape, mean, stdev = t.fit(scores_)
kstat, kval = stats.kstest(scores_, 't', args=[shape, mean, stdev])
print('kstat:', kstat, ' kval:', kval)
```

```python
fig, ax = plt.subplots(1, 1)
x = np.linspace(t.ppf(0.025, shape, mean, stdev),
                t.ppf(0.975, shape, mean, stdev), 1000)
ax.plot(x, t.pdf(x, shape, mean, stdev), 'r-', lw=5,
        alpha=0.6, label='t dist. pdf')
ax.hist(scores_, density=True, histtype='stepfilled',
        alpha=0.6, bins=60)
ax.legend(loc='best', frameon=False)
plt.show()
```
