```python
import pickle
from plotnine import * 
from pandas import * 
```

```python
data=pickle.load(open("AD_CTRL_SMTG.preds.0",'rb'))
```

```python
preds=np.concatenate((data[0],data[1].values),axis=1)
```

```python
preds_df=DataFrame(data=preds,
                  index=range(preds.shape[0]),
                  columns=["predicted","labels"])
```

```python
preds_df.head()
```

```python
nonzero_preds_df=preds_df[preds_df['labels']>0]
```

```python
nonzero_preds_df.head()
```

```python
ggplot(nonzero_preds_df,aes(x="labels",y="predicted"))+geom_point(size=1,alpha=0.1)+xlab("Labels")+ylab("Predicted")+ggtitle("AD_CTRL_SMTG.preds.0")
```
